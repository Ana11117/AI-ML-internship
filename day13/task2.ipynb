```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'SquareFootage': [1200, 1500, 1800, 2000, 2200, 2500, 2700, 3000, 3200, 3500],
    'Price': [200000, 250000, 280000, 310000, 330000, 400000, 420000, 480000, 500000, 550000],
    'Location': ['Urban', 'Urban', 'Suburban', 'Suburban', 'Rural',
                 'Urban', 'Suburban', 'Urban', 'Rural', 'Suburban']
}

df = pd.DataFrame(data)

df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))
sns.scatterplot(x='SquareFootage', y='Price', data=df)

plt.title("Square Footage vs Price")
plt.xlabel("Square Footage")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Location', y='Price', data=df)

plt.title("Location vs Price")
plt.xlabel("Location")
plt.ylabel("Price")
plt.show()
```
