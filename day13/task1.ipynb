```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r"C:\Users\DELL\Downloads\housing.csv")
```

```python
data.head()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data['median_house_value'], bins=30, kde=True)
plt.title("Distribution of Median House Value")
plt.xlabel("Median House Value")
plt.ylabel("Frequency")
plt.show()
```

```python
skewness = data['median_house_value'].skew()
kurtosis = data['median_house_value'].kurt()

print("Skewness:", skewness)
print("Kurtosis:", kurtosis)
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='ocean_proximity', data=data)
plt.title("Count of Houses by Ocean Proximity")
plt.xticks(rotation=45)
plt.show()
```

```python
data['log_house_value'] = np.log(data['median_house_value'])
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data['log_house_value'], bins=30, kde=True)
plt.title("Distribution After Log Transformation")
plt.xlabel("Log House Value")
plt.ylabel("Frequency")
plt.show()
```

```python
print("Skewness After Log:", data['log_house_value'].skew())
```
