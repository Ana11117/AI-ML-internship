```python
import sqlite3
import pandas as pd
conn = sqlite3.connect("internship.db")
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS interns (
    id INTEGER PRIMARY KEY,
    name TEXT,
    track TEXT,
    stipend INTEGER
)
""")

conn.commit()
```

```python
intern_data = [
    (1, "Aisha", "Data Science", 7000),
    (2, "Rahul", "Web Dev", 5000),
    (3, "Sneha", "Data Science", 6000),
    (4, "Arjun", "Web Dev", 4500),
    (5, "Meera", "Data Science", 8000)
]

cursor.executemany("INSERT OR REPLACE INTO interns VALUES (?, ?, ?, ?)", intern_data)
conn.commit()
```

```python
query_filter = """
SELECT *
FROM interns
WHERE track = 'Data Science'
AND stipend > 5000;
"""

df_filter = pd.read_sql_query(query_filter, conn)
df_filter
```

```python
query_avg = """
SELECT track, AVG(stipend) AS average_stipend
FROM interns
GROUP BY track;
"""

df_avg = pd.read_sql_query(query_avg, conn)
df_avg
```

```python
query_count = """
SELECT track, COUNT(*) AS total_interns
FROM interns
GROUP BY track;
"""

df_count = pd.read_sql_query(query_count, conn)
df_count
```

```python
conn.close()
```

```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("internship.db")
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS interns (
    id INTEGER PRIMARY KEY,
    name TEXT,
    track TEXT,
    stipend INTEGER
)
""")
conn.commit()
```

```python
cursor.execute("DELETE FROM interns")  # optional: clears old data

intern_data = [
    (1, "Aisha", "Data Science", 15000),
    (2, "Rohan", "Web Dev", 12000),
    (3, "Megha", "AI", 18000),
    (4, "Kiran", "Cyber Security", 16000),
    (5, "Sara", "Data Science", 17000)
]

cursor.executemany("INSERT INTO interns VALUES (?, ?, ?, ?)", intern_data)
conn.commit()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS mentors (
    mentor_id INTEGER PRIMARY KEY,
    mentor_name TEXT,
    track TEXT
)
""")
conn.commit()
```

```python
cursor.execute("DELETE FROM mentors")  # optional

mentor_data = [
    (1, "Rahul", "Data Science"),
    (2, "Sneha", "Web Dev"),
    (3, "Arjun", "AI"),
    (4, "Meera", "Cyber Security")
]

cursor.executemany("INSERT INTO mentors VALUES (?, ?, ?)", mentor_data)
conn.commit()
```

```python
query = """
SELECT interns.name, interns.track, mentors.mentor_name
FROM interns
INNER JOIN mentors
ON interns.track = mentors.track;
"""

df = pd.read_sql_query(query, conn)

df
```
