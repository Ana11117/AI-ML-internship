```python
Q1 = data['Age'].quantile(0.25)
Q3 = data['Age'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

data = data[(data['Age'] >= lower) & (data['Age'] <= upper)]
```

```python
Q1 = data['AnnualIncome'].quantile(0.25)
Q3 = data['AnnualIncome'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

data = data[(data['AnnualIncome'] >= lower) & (data['AnnualIncome'] <= upper)]
```

```python
sns.boxplot(data=data,orient="y")
```
